# CLEANUP
Using  
WHEN
AS
CASE
SQL keywords

```python
c.execute( '''
DROP TABLE IF EXISTS customer_orders_temp;

''')


c.execute( '''
CREATE TABLE customer_orders_temp AS
  SELECT
    order_id,
    customer_id,
    pizza_id,
    exclusions,

    CASE
      WHEN extras = 'null'  OR  extras is null THEN ''
      ELSE extras
    END AS extras,


    order_time

FROM customer_orders;
        ''')
```

```python

pd.read_sql("SELECT *  FROM customer_orders_temp;", conn)
```

```python
c.execute( '''
DROP TABLE IF EXISTS customer_orders_temp;

''')


c.execute( '''
CREATE TABLE customer_orders_temp AS
  SELECT
    order_id,
    customer_id,
    pizza_id,
    CASE
      WHEN exclusions = 'null'  OR  exclusions is null THEN ''
      ELSE exclusions
    END AS exclusions,

    CASE
      WHEN extras = 'null'  OR  extras is null THEN ''
      ELSE extras
    END AS extras,


    order_time

FROM customer_orders;
        ''')
```

```python
## JIC

c.execute( '''
DROP TABLE IF EXISTS runner_orders_temp;

''')
c.execute('''
 CREATE TEMP TABLE runner_orders_temp AS
 SELECT
   order_id,
   runner_id,
   CASE
    WHEN pickup_time = 'null'  OR  pickup_time is null THEN ''
     ELSE pickup_time
   END AS pickup_time,


    CASE
    WHEN distance = 'null'  THEN ''
    WHEN distance LIKE '%km' THEN SUBSTR(distance, 1, LENGTH(distance) - 2)



    ELSE distance
     END AS distance ,


    CASE
     WHEN duration = 'null' OR duration is null THEN ''
     WHEN duration LIKE '%mins' THEN  REPLACE(duration, 'mins', '')
     WHEN duration LIKE '%minute'THEN REPLACE(duration, 'minute', '')
     WHEN duration LIKE '%minutes'THEN REPLACE(duration, 'minutes', '')
     ELSE duration
   END AS duration,


    CASE
     WHEN cancellation IS NULL or cancellation = 'null' THEN ''
     ELSE cancellation
   END AS cancellation

    FROM runner_orders

''')
```
